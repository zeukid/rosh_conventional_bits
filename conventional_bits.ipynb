```python
N = 77
a = 8
```

```python
z = list(range(N))
y = [a**z0 % N for z0 in z]

mpl.plot(z,y)
mpl.xlabel('z')
mpl.ylabel(f'{a}^z (mod {N})')
mpl.show()
```

```python
r = z[y[1:].index(1)+1]
print(r)
```

```python
math.gcd(r-1,N)
```
